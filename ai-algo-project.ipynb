```python
import time
import matplotlib.pyplot as plt
```

```python
file_path = '/kaggle/input/temperaturee/temperature_dataset.txt'
with open(file_path, 'r') as file:
    # Read the file, remove newline characters, and split by ','
    data_str = file.read().replace('\n', '').strip('[]').split(', ')

    # Convert strings to float
    array = [float(value) for value in data_str]

print(array)
```

```python
def binary_search(arr, target):
    start_time = time.time()
    low, high = 0, len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        mid_value = arr[mid]

        if mid_value == target:
            end_time = time.time()
            elapsed_time = end_time - start_time
            return mid, elapsed_time
        elif mid_value < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1, 0  
```

```python
sorted_array = sorted(array)
target_value = 10.715632427863287
result = binary_search(sorted_array, target_value)

if result[0] != -1:
    print(f'Target {target_value} found at index {result[0]}')
else:
    print(f'Target {target_value} not found in the array')

binary_search_result, binary_search_time = result

if binary_search_result != -1:
    print(f'Binary Search: Target {target_value} found at index {binary_search_result} in {binary_search_time} seconds')
else:
    print(f'Binary Search: Target {target_value} not found in the array in {binary_search_time} seconds')
```

```python
def linear_search(arr, target):
    start_time = time.time()
    for i, value in enumerate(arr):
        if value == target:
            end_time = time.time()
            elapsed_time = end_time - start_time
            return i, elapsed_time
    return -1, 0  
```

```python
sorted_array = sorted(array)
target_value = 10.715632427863287
result = linear_search(sorted_array, target_value)

if result[0] != -1:
    print(f'Target {target_value} found at index {result[0]}')
else:
    print(f'Target {target_value} not found in the array')

linear_search_result, linear_search_time = result

if linear_search_result != -1:
    print(f'Linear Search: Target {target_value} found at index {linear_search_result} in {linear_search_time} seconds')
else:
    print(f'Linear Search: Target {target_value} not found in the array in {linear_search_time} seconds')
```

```python
# Plotting
labels = ['Binary Search', 'Linear Search']
times = [binary_search_time, linear_search_time]
plt.bar(labels, times)
plt.ylabel('Time (seconds)')
plt.title('Comparison of Search Algorithms')
plt.show()
```
